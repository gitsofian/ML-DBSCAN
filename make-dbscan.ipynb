#   make_blobs : DBSCAN

```python
from sre_constants import NOT_LITERAL_UNI_IGNORE
from sklearn.datasets import make_blobs, make_moons, make_circles
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN


X, y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=0)

# Plotting data
plt.figure(1)
plt.scatter(X[:, 0], X[:, 1], c=y, marker=".")
plt.title("Ground - Datasets Truth (make blobs)")

# Erstelle DBSCAN
eps=0.5
min_samples = 5
for eps in range(1, 9):
    eps /= 10
    dbscan_clustering = DBSCAN(eps=eps, min_samples=min_samples, algorithm='ball_tree').fit(X)
    y_clustering = dbscan_clustering.fit_predict(X)
    print(f"eps, min_samples, labels: \n eps: {eps},\t min samples: {min_samples},\t labels: {np.unique(dbscan_clustering.labels_)}\n")

# print(f"K-Means labels Attribute: \n{kmeans.labels_}")
# print(f"K-Means Cluster Center Attribute: \n{kmeans.cluster_centers_}")

plt.figure(2)
plt.scatter(X[:, 0], X[:, 1], c=y_clustering, marker=".")
# plt.scatter(y_clustering ,marker="D", linewidths=2, c='#ff0000')
plt.title("DBSCAN Clustering")

# Add Axes Labels
plt.xlabel("x1 axis")
plt.ylabel("x2 axis")

plt.show()
```

# make_moons : DBSCAN

```python
from sre_constants import NOT_LITERAL_UNI_IGNORE
from sklearn.datasets import make_blobs, make_moons, make_circles
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN


X, y = make_moons(n_samples=(1000,1000), noise=0.1, random_state=0)

# Plotting data
plt.figure(1)
plt.scatter(X[:, 0], X[:, 1], c=y, marker=".")
plt.title("Ground - Datasets Truth (make moons)")

# Erstelle DBSCAN
dbscan_clustering = DBSCAN(eps=.3, min_samples=5).fit(X)
y_clustering = dbscan_clustering.fit_predict(X)

# print(f"K-Means labels Attribute: \n{kmeans.labels_}")
# print(f"K-Means Cluster Center Attribute: \n{kmeans.cluster_centers_}")


plt.figure(2)
plt.scatter(X[:, 0], X[:, 1], c=y_clustering, marker=".")
# plt.scatter(y_clustering ,marker="D", linewidths=2, c='#ff0000')
plt.title("DBSCAN Clustering")

# Add Axes Labels
plt.xlabel("x1 axis")
plt.ylabel("x2 axis")

plt.show()
```
